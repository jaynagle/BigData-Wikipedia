```python
import nltk
from nltk.corpus import wordnet
from nltk.corpus import wordnet as wn
from os import listdir
```

```python
# Open a file
import os, sys
path = "C:\Second Sem\Big Data and Analytics\Project\Manually Extracted Text Files"
dirs = os.listdir( path )
print (dirs)
# This would print all the files and directories
```

```python
fileToNoun={}
for file in dirs:
    files= open(path + "\\" + file,"r")
    reading= files.readlines()
    files.close()
    ignoreCount = 0
    #print (reading)
    noun=[]
    verb=[]
    #print(reading)
    for i in reading:
        try:
            i.strip()
            n=str(i)
            #text = n.split()
            text=nltk.word_tokenize(n)
            k=nltk.pos_tag(text)
            for j in k:
                if (len(j[0])<=2) or j[0]=="edit":
                    continue
                elif j[1]=="NN":
                    #print "appending" + j[0]
                    noun.append(j[0])
        except Exception,e:
            #print str(e)
            #traceback.print_exc()
            ignoreCount = ignoreCount + 1
    fileToNoun[file] = noun
    #print str(file) + str(noun)
    print "Words Ignored: ", ignoreCount
print "done"
#nfreq=nltk.FreqDist(noun)
#vfreq=nltk.FreqDist(verb)

```
